```python
df = pd.read_csv('/home/cristiano/www/skin_cancer_prediction/ISIC_2020/ISIC_2020_label_imgs.csv')
```

```python
df.head()
```

```python
mel_list = df[df['MEL']==1.0]['image'].tolist()
len(mel_list)
```

```python
source = '/home/cristiano/www/skin_cancer_prediction/ISIC_2020/ISIC_2019_dataset_Input/'
destination = '/home/cristiano/www/skin_cancer_prediction/ISIC_2020/division/imgs_melanoma/training/'

allfiles = os.listdir(source)

for f in allfiles:
    m=f.split('.')[0]
    if m in mel_list:
        file="melanoma_"+str(f)
        print(file)
        shutil.copy(source+f, destination+file)

print("Imagens do melanoma rotuladas com sucesso.")
```
