```python
import os
import random
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import KFold
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint
import gc
import warnings

warnings.filterwarnings("ignore", category=UserWarning, module='keras.src.trainers.data_adapters.py_dataset_adapter')

# Bloco 1: Preparação dos dados
B_DIR = "/home/cristiano/www/skin_cancer_prediction/ISIC_2020/division_imgs/carcinoma_basocelular/train/"
M_DIR = "/home/cristiano/www/skin_cancer_prediction/ISIC_2020/division_imgs/melanoma/train/"
R_DIR = "/home/cristiano/www/skin_cancer_prediction/ISIC_2020/division_imgs/random_images/train"  # Diretório para imagens aleatórias

def visualize_samples(image_dir, title, sample_size=6):
    image_files = os.listdir(image_dir)
    samples = random.sample(image_files, sample_size)
    f, ax = plt.subplots(2, 3, figsize=(15, 9))
    for i in range(sample_size):
        im = cv2.imread(os.path.join(image_dir, samples[i]))
        ax[i // 3, i % 3].imshow(im)
        ax[i // 3, i % 3].axis('off')
    f.suptitle(title, fontsize=20)
    plt.show()

def load_images_from_directory(directory, label, image_size=(224, 224)):
    data = []
    labels = []
    image_files = os.listdir(directory)
    for image_file in image_files:
        image_path = os.path.join(directory, image_file)
        image = cv2.imread(image_path)
        if image is not None:
            image_from_array = Image.fromarray(image, 'RGB')
            resized_image = image_from_array.resize(image_size)
            data.append(np.array(resized_image))
            labels.append(label)
    return data, labels

visualize_samples(B_DIR, 'Carcinoma Basocelular')
visualize_samples(M_DIR, 'Melanoma')
visualize_samples(R_DIR, 'Random Images')  # Visualiza amostras de imagens aleatórias

# Carrega e prepara dados
basal_data, basal_labels = load_images_from_directory(B_DIR, 0)
melanoma_data, melanoma_labels = load_images_from_directory(M_DIR, 1)
random_data, random_labels = load_images_from_directory(R_DIR, 2)  # Nova classe para imagens aleatórias

# Balanceamento das classes
min_class_size = min(len(basal_data), len(melanoma_data), len(random_data))
basal_data = basal_data[:min_class_size]
basal_labels = basal_labels[:min_class_size]
melanoma_data = melanoma_data[:min_class_size]
melanoma_labels = melanoma_labels[:min_class_size]
random_data = random_data[:min_class_size]
random_labels = random_labels[:min_class_size]

# Combina e embaralha os dados
data = np.array(basal_data + melanoma_data + random_data)
labels = np.array(basal_labels + melanoma_labels + random_labels)
shuffled_indices = np.arange(len(data))
np.random.shuffle(shuffled_indices)
data = data[shuffled_indices]
labels = labels[shuffled_indices]

# Salva arrays
np.save("/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/feats_train_up", data)
np.save("/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/labels_train_up", labels)

# Bloco 2: treinamento e avaliação com o KFold

data = np.load("/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/feats_train_up.npy")
labels = np.load("/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/labels_train_up.npy")

# Divide dados em treinamento e teste (80% para o KFold, 20% para teste nos 5 modelos)
split_index = int(0.8 * len(data))
x_train, x_test = data[:split_index], data[split_index:]
y_train, y_test = labels[:split_index], labels[split_index:]

# Normaliza os dados
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
num_classes = len(np.unique(labels))
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)

trainAug = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=False,
    vertical_flip=False
)

# Criação do modelo
def create_model():
    base_model = tf.keras.applications.MobileNetV2(
        weights='imagenet', 
        include_top=False, 
        input_shape=(224, 224, 3)
    )
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.2)(x)
    predictions = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Configuração callbacks para salvar o melhor modelo em cada fold
def create_callbacks():
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=2, verbose=1, mode='min', min_delta=1e-10)
    checkpoint = ModelCheckpoint("best_model_fold.keras", monitor='val_loss', verbose=1, save_best_only=True, mode='min')
    return [reduce_lr, checkpoint]

# Plotagem da matriz de confusão
def plot_confusion_matrix(cm, classes, fold=''):
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
    plt.ylabel('Rótulo verdadeiro')
    plt.xlabel('Rótulo previsto')
    plt.title(f'Matriz de Confusão {fold}')
    plt.show()

# Cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

best_val_accuracies = []
best_f1_scores = []
best_model_paths = []

for fold, (train_index, val_index) in enumerate(kf.split(x_train)):
    print(f"Treinando o fold {fold + 1}/{kf.n_splits}...")
    x_train_fold, x_val_fold = x_train[train_index], x_train[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    model = create_model()
    
    # Treina o modelo
    model.fit(
        trainAug.flow(x_train_fold, y_train_fold, batch_size=16),
        validation_data=(x_val_fold, y_val_fold),
        epochs=30,
        callbacks=create_callbacks()
    )

    # Avalia o modelo
    val_loss, val_accuracy = model.evaluate(x_val_fold, y_val_fold, verbose=0)
    print(f'Fold {fold + 1} - Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}')

    # Plota matriz de confusão para o fold atual
    y_val_pred = model.predict(x_val_fold, batch_size=16)
    y_val_pred_classes = np.argmax(y_val_pred, axis=1)
    y_val_true = np.argmax(y_val_fold, axis=1)
    confusion_mtx = confusion_matrix(y_val_true, y_val_pred_classes)
    plot_confusion_matrix(confusion_mtx, classes=range(num_classes), fold=f'Fold {fold + 1}')

    # Calcula o F1-score para o fold atual
    report = classification_report(y_val_true, y_val_pred_classes, output_dict=True, target_names=['BCC', 'Melanoma', 'Aleatória'])
    f1_score = (report['BCC']['f1-score'] + report['Melanoma']['f1-score'] + report['Aleatória']['f1-score']) / 3

    best_val_accuracies.append(val_accuracy)
    best_f1_scores.append(f1_score)
    
    # Salva o melhor modelo
    best_model_path = f'/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/skin_model_keras/epoch_up/best_model_fold_{fold + 1}.keras'
    model.save(best_model_path)
    best_model_paths.append(best_model_path)

    # Limpa memória
    del x_train_fold, x_val_fold, y_train_fold, y_val_fold, model
    gc.collect()
    tf.keras.backend.clear_session()

# Bloco 3: Avaliação dos Modelos com Dados de Teste

all_test_accuracies = []
all_test_f1_scores = []

for i, model_path in enumerate(best_model_paths):
    print(f"Avaliando o modelo do fold {i + 1}...")
    model = load_model(model_path)
    test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
    
    y_test_pred = model.predict(x_test)
    y_test_pred_classes = np.argmax(y_test_pred, axis=1)
    y_test_true = np.argmax(y_test, axis=1)

    report = classification_report(y_test_true, y_test_pred_classes, output_dict=True, target_names=['BCC', 'Melanoma', 'Aleatória'])
    f1_score = (report['BCC']['f1-score'] + report['Melanoma']['f1-score'] + report['Aleatória']['f1-score']) / 3
    
    print(f'Teste do fold {i + 1} - Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}')
    all_test_accuracies.append(test_accuracy)
    all_test_f1_scores.append(f1_score)

    # Plota matriz de confusão para o fold atual nos dados de teste
    confusion_mtx = confusion_matrix(y_test_true, y_test_pred_classes)
    plot_confusion_matrix(confusion_mtx, classes=range(num_classes), fold=f'Teste Fold {i + 1}')

# Bloco 4: Média dos Resultados e Melhor Modelo

avg_test_accuracy = np.mean(all_test_accuracies)
avg_test_f1_score = np.mean(all_test_f1_scores)

print(f"Acurácia Média dos Modelos: {avg_test_accuracy:.4f}")
print(f"F1-Score Médio dos Modelos: {avg_test_f1_score:.4f}")

best_model_index = np.argmax(all_test_accuracies)
print(f"O melhor modelo é o do fold {best_model_index + 1} com acurácia de {all_test_accuracies[best_model_index]:.4f}.")

```

```python
import tensorflow as tf

# Carrega o modelo Keras treinado (neste caso, o melhor modelo de um fold específico)
best_model_path = f'/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/01-mobileNetV2/fold_up/best_model_fold_4.keras'
model = tf.keras.models.load_model(best_model_path)

# Converte o modelo para TFLite
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# Opcional: Aplicar otimizações (quantização)
# Isso reduz o tamanho do modelo e melhora a performance, útil para dispositivos móveis
converter.optimizations = [tf.lite.Optimize.DEFAULT]

# Converte o modelo para o formato TFLite
tflite_model = converter.convert()

# Salva o modelo convertido
tflite_model_path = f'/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/01-mobileNetV2/fold_up/best_model_fold_4.tflite'
with open(tflite_model_path, 'wb') as f:
    f.write(tflite_model)

print(f"Modelo TFLite salvo em: {tflite_model_path}")
```
