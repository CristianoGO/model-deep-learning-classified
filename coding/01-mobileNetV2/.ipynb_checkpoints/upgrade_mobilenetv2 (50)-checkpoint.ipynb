```python
# Bloco 01: bibliotecas e preparao dos dados

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import random
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import KFold
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint
import gc
import warnings

warnings.filterwarnings("ignore", category=UserWarning, module='keras.src.trainers.data_adapters.py_dataset_adapter')

B_DIR = "/home/cristiano/www/skin_cancer_prediction/ISIC_2020/division_imgs/carcinoma_basocelular/train/"
M_DIR = "/home/cristiano/www/skin_cancer_prediction/ISIC_2020/division_imgs/melanoma/train/"

def visualize_samples(image_dir, title, sample_size=6):
    image_files = os.listdir(image_dir)
    samples = random.sample(image_files, sample_size)
    f, ax = plt.subplots(2, 3, figsize=(15, 9))
    for i in range(sample_size):
        im = cv2.imread(os.path.join(image_dir, samples[i]))
        ax[i // 3, i % 3].imshow(im)
        ax[i // 3, i % 3].axis('off')
    f.suptitle(title, fontsize=20)
    plt.show()

def load_images_from_directory(directory, label, image_size=(224, 224)):
    data = []
    labels = []
    image_files = os.listdir(directory)
    for image_file in image_files:
        image_path = os.path.join(directory, image_file)
        image = cv2.imread(image_path)
        if image is not None:
            image_from_array = Image.fromarray(image, 'RGB')
            resized_image = image_from_array.resize(image_size)
            data.append(np.array(resized_image))
            labels.append(label)
    return data, labels

# Visualizando amostras das imagens
visualize_samples(B_DIR, 'Carcinoma Basocelular')
visualize_samples(M_DIR, 'Melanoma')

basal_data, basal_labels = load_images_from_directory(B_DIR, 0)
melanoma_data, melanoma_labels = load_images_from_directory(M_DIR, 1)

# Combina e embaralha os dados
data = np.array(basal_data + melanoma_data)
labels = np.array(basal_labels + melanoma_labels)

# Balanceamento dos dados para ter o mesmo nmero de imagens para cada classe
min_samples = min(np.sum(labels == 0), np.sum(labels == 1))
data_balanced = []
labels_balanced = []

for label in np.unique(labels):
    class_indices = np.where(labels == label)[0]
    selected_indices = np.random.choice(class_indices, min_samples, replace=False)
    data_balanced.extend(data[selected_indices])
    labels_balanced.extend(labels[selected_indices])

data = np.array(data_balanced)
labels = np.array(labels_balanced)

# Salva arrays
np.save("/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/feats_train_50", data)
np.save("/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/labels_train_50", labels)
```

```python
# Bloco 02: aplicando as funes e configuraes do modelo

def create_model():
    base_model = tf.keras.applications.MobileNetV2(
        weights='imagenet', 
        include_top=False, 
        input_shape=(224, 224, 3)
    )
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.2)(x)
    predictions = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def create_callbacks(fold):
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=2, verbose=1, mode='min', min_delta=1e-10)
    checkpoint = ModelCheckpoint(f"best_model_fold_{fold}.keras", monitor='val_loss', verbose=1, save_best_only=True, mode='min')
    return [reduce_lr, checkpoint]

def plot_confusion_matrix(cm, classes):
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
    plt.ylabel('Rtulo verdadeiro')
    plt.xlabel('Rtulo previsto')
    plt.title('Matriz de confuso')
```

```python
# Bloco 03: treinamento e avaliao do cross-validation

# Normaliza os dados
data = data.astype('float32') / 255
num_classes = len(np.unique(labels))
labels = to_categorical(labels, num_classes)

# Augmentao de dados
trainAug = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=False,
    vertical_flip=False
)

# Cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

fold_results = []
best_val_accuracy = 0.0
best_model_path = None

for fold, (train_index, val_index) in enumerate(kf.split(data)):
    print(f"Treinando o fold {fold + 1}/{kf.n_splits}...")
    x_train_fold, x_val_fold = data[train_index], data[val_index]
    y_train_fold, y_val_fold = labels[train_index], labels[val_index]

    # Cria um novo modelo para cada fold
    model = create_model()
    
    # Treina o modelo
    model.fit(
        trainAug.flow(x_train_fold, y_train_fold, batch_size=16),
        validation_data=(x_val_fold, y_val_fold),
        epochs=50,
        callbacks=create_callbacks(fold + 1)
    )

    # Avalia o modelo
    val_loss, val_accuracy = model.evaluate(x_val_fold, y_val_fold, verbose=0)
    print(f'Fold {fold + 1} - Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}')

    # Armazena os resultados do fold
    y_val_pred = model.predict(x_val_fold, batch_size=16)
    y_val_pred_classes = np.argmax(y_val_pred, axis=1)
    y_val_true = np.argmax(y_val_fold, axis=1)
    confusion_mtx = confusion_matrix(y_val_true, y_val_pred_classes)
    
    # Salva o relatrio de classificao
    report = classification_report(y_val_true, y_val_pred_classes, target_names=['BCC', 'Melanoma'])
    fold_results.append({
        'fold': fold + 1,
        'accuracy': val_accuracy,
        'classification_report': report
    })
    
    # Plota matriz de confuso para o fold atual
    plot_confusion_matrix(confusion_mtx, classes=['BCC', 'Melanoma'])
    plt.show()

    if val_accuracy > best_val_accuracy:
        best_val_accuracy = val_accuracy
        best_model_path = f'/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/skin_model_keras/epoch_50/best_model_fold_{fold + 1}.keras'
        model.save(best_model_path)

    # Limpa memria
    del x_train_fold, x_val_fold, y_train_fold, y_val_fold, model
    gc.collect()
    tf.keras.backend.clear_session()
```

```python
# Bloco 04: avaliao do melhor modelo, resultados dos folds, mdia de acurcia e f1-score

# Carrega o melhor modelo treinado durante o cross-validation
if best_model_path:
    model = load_model(best_model_path)
    print(f"Melhor modelo carregado de: {best_model_path}")
else:
    raise ValueError("Nenhum modelo foi salvo durante o treinamento!")

# Avaliao com os dados de teste
batch_size = 16
num_samples = data.shape[0]
num_batches = int(np.ceil(num_samples / batch_size))

pred_classes = []
true_classes = []

for i in range(num_batches):
    start = i * batch_size
    end = min((i + 1) * batch_size, num_samples)
    x_batch = data[start:end]
    y_batch = labels[start:end]

    pred_Y_batch = model.predict(x_batch, batch_size=batch_size, verbose=0)
    pred_classes_batch = np.argmax(pred_Y_batch, axis=1)
    true_classes_batch = np.argmax(y_batch, axis=1)
    
    pred_classes.extend(pred_classes_batch)
    true_classes.extend(true_classes_batch)

# Relatrio de Classificao para o Melhor Modelo
print("\nRelatrio de Classificao para o Melhor Modelo:")
print(classification_report(true_classes, pred_classes, target_names=['BCC', 'Melanoma']))

# Exibe os resultados de cada fold
for result in fold_results:
    print(f"\nResultados para o Fold {result['fold']}:")
    print(f"Acurcia: {result['accuracy']:.4f}")
    print("Relatrio de Classificao:")
    print(result['classification_report'])

# Mdia das acurcias e F1-scores dos folds
accuracy_mean = np.mean([result['accuracy'] for result in fold_results])

# Extrao dos F1-scores para cada classe e clculo da mdia
f1_scores = []
for result in fold_results:
    report_lines = result['classification_report'].split('\n')
    for line in report_lines:
        if 'BCC' in line or 'Melanoma' in line:
            f1_score = float(line.split()[-2])
            f1_scores.append(f1_score)

f1_score_mean = np.mean(f1_scores)

print(f'\nMdia da Acurcia dos Folds: {accuracy_mean:.4f}')
print(f'Mdia do F1-Score dos Folds: {f1_score_mean:.4f}')
```
