# Divisão de treinamento e teste

```python
# Divide as imagens de células na proporção de 80:20, ou seja, 80% para treinamento e 20% para fins de teste
(x_train,x_test)=feats[(int)(0.2*len_data):],feats[:(int)(0.2*len_data)]
(y_train,y_test)=labels[(int)(0.2*len_data):],labels[:(int)(0.2*len_data)]
```

# Normalização de dados de imagem

```python
# Como estamos trabalhando nos dados da imagem, normalizamos os dados dividindo 255.
x_train = x_train.astype('float32')/255
x_test = x_test.astype('float32')/255
train_len=len(x_train)
test_len=len(x_test)
```

```python
y_train=to_categorical(y_train,3)
y_test=to_categorical(y_test,3)
```

# Processamento das imagens (redimensionamento, qualidade)

```python
trainAug = ImageDataGenerator(
    featurewise_center=False,  # Define a média de entrada como 0 sobre o conjunto de dados
    samplewise_center=False,  # Define a média de cada amostra como 0
    featurewise_std_normalization=False,  # Divide as entradas pelo desvio padrão do conjunto de dados
    samplewise_std_normalization=False,  # Divide cada entrada pelo seu desvio padrão
    zca_whitening=False,  # Aplicar branqueamento ZCA
    rotation_range=10,  # Rotacionar aleatoriamente imagens no intervalo (graus, 0 a 180)
    zoom_range=0.1,  # Zoom aleatório na imagem
    width_shift_range=0.1,  # Deslocar aleatoriamente imagens horizontalmente (fração da largura total)
    height_shift_range=0.1,  # Deslocar aleatoriamente imagens verticalmente (fração da altura total)
    horizontal_flip=False,  # Inverter aleatoriamente imagens horizontalmente
    vertical_flip=False)  # Inverter aleatoriamente imagens verticalmente
```

# Construção de modelo

```python
conv_base = MobileNetV2(
    include_top=False,
    input_shape=(224, 224, 3),
    weights='imagenet')

for layer in conv_base.layers:
    layer.trainable = True
```

```python
from tensorflow.keras import layers
from tensorflow.keras.models import Model

x = conv_base.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(64, activation='relu')(x)
x = layers.Dropout(0.1)(x)
predictions = layers.Dense(3, activation='softmax')(x)
model = Model(conv_base.input, predictions)
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.models import load_model

# Carregar o modelo pré-treinado
model = load_model('.mdl_wts.keras')

# Retornos de chamada (Callbacks)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=2, verbose=1, mode='min', min_delta=0.00000000001)
checkpoint = ModelCheckpoint("best_model_{epoch:02d}.keras", monitor='val_loss', verbose=1, save_best_only=False, mode='min')
callbacks = [reduce_lr, checkpoint]

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
BS = 64

print("Treinando o modelo...")
total_train_steps = len(x_train) // BS  # Número de etapas com base no tamanho do conjunto de treinamento

H = model.fit(
    x_train, y_train,
    batch_size=BS,
    shuffle=True,
    validation_data=(x_test, y_test),
    epochs=30,
    callbacks=callbacks)

model.save('/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/skin_model_keras/model_v1.keras')
```

# Plotando a Área Sob A Curva Característica de Operação do Receptor - ROC AUC

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_classification
from sklearn.preprocessing import label_binarize

# Bibliotecas relacionadas à análise de curvas ROC
from sklearn.metrics import roc_curve, roc_auc_score, auc
from scipy import interpolate
from itertools import cycle

y_test = np.array(y_test)

n_classes = 3

pred_Y = model.predict(x_test, batch_size = 16, verbose = True)

lw = 2

# Calcula a curva ROC e a área ROC para cada classe
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], pred_Y[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    
# Calcular curva ROC micromédia e área ROC
fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), pred_Y.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Gráfico de uma curva ROC para uma classe específica
for i in range(n_classes):
    plt.figure()
    plt.plot(fpr[i], tpr[i], label='Curva ROC (área = %0.2f)' % roc_auc[i])
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Taxa do falso positivo')
    plt.ylabel('Taxa do verdadeiro positivo')
    plt.title('Característica operacional do receptor')
    plt.legend(loc="lower right")
    plt.show()

# Primeiro agregue todas as taxas de falsos positivos
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

# Em seguida, interpole todas as curvas ROC nestes pontos
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

# Calculando a média e AUC
mean_tpr /= n_classes

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# Plota todas as curvas ROC
fig = plt.figure(figsize=(12, 8))
plt.plot(fpr["micro"], tpr["micro"],
         label='Curva ROC micromédia (área = {0:0.2f})'
               ''.format(roc_auc["micro"]),
         color='deeppink', linestyle=':', linewidth=4)

plt.plot(fpr["macro"], tpr["macro"],
         label='Curva ROC macromédia (área = {0:0.2f})'
               ''.format(roc_auc["macro"]),
         color='navy', linestyle=':', linewidth=4)

colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=lw,
             label='Curva ROC da classe {0} (área = {1:0.2f})'
             ''.format(i, roc_auc[i]))


plt.plot([0, 1], [0, 1], 'k--', lw=lw)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa do falso positivo')
plt.ylabel('Taxa do verdadeiro positivo')
plt.title('Característica operacional do receptor para multiclasse')
plt.legend(loc="lower right")
sns.despine()
plt.show()
```

# Teste 01 com amostras de imagens

```python
from keras.preprocessing import image
import numpy as np
from PIL import Image

img1 ="/home/cristiano/www/skin_cancer_prediction/ISIC_2020/division_imgs/cercinoma_basocelular/test/bcc_ISIC_0025584.jpg"
img2 = "/home/cristiano/www/skin_cancer_prediction/ISIC_2020/division_imgs/melanoma/test/melanoma_ISIC_0026158.jpg"
class_dict = {0: "Basal_Cell_Carcinoma (Cancer)", 1: "Melanoma (Cancer)"}

def pred_image(img_path, model):
    img = image.load_img(img_path, target_size=(224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img.astype('float32') / 255

    preds = model.predict(img)
    pred = np.argmax(preds, axis=1)
    confidence = np.max(preds) * 100
    
    if pred[0] in class_dict:
        pred_cat = class_dict[pred[0]]
        return pred_cat, confidence
    else:
        return "Classe prevista não reconhecida", confidence

result1 = pred_image(img1, model)
result2 = pred_image(img2, model)

print("Resultado para a imagem 1:")
print("  - Categoria prevista:", result1[0])
print("  - Confiança:", result1[1], "%")
print()

print("Resultado para a imagem 2:")
print("  - Categoria prevista:", result2[0])
print("  - Confiança:", result2[1], "%")
print()
```
