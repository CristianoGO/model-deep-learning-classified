```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import KFold
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import tensorflow as tf
import gc

def create_model(num_classes):
    base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    predictions = Dense(num_classes, activation='softmax')(x)

    model = tf.keras.Model(inputs=base_model.input, outputs=predictions)

    for layer in base_model.layers:
        layer.trainable = False

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def create_callbacks():
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    model_checkpoint = ModelCheckpoint(filepath='best_model.keras', monitor='val_accuracy', save_best_only=True)
    return [early_stopping, model_checkpoint]

def plot_confusion_matrix(cm, classes, fold, title='Matriz de Confusão', cmap=plt.cm.Blues):
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(f'{title} - {fold}')
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in np.ndindex(cm.shape):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('Classe Verdadeira')
    plt.xlabel('Classe Predita')
    plt.tight_layout()
    plt.show()

# Definindo variáveis e gerador de dados
num_classes = 3
image_size = (224, 224)
batch_size = 16
num_epochs = 30

# Geradores de dados
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
train_generator = train_datagen.flow_from_directory(
    '/home/cristiano/www/skin_cancer_prediction/ISIC_2020/division_imgs/',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)
val_generator = train_datagen.flow_from_directory(
    '/home/cristiano/www/skin_cancer_prediction/ISIC_2020/division_imgs/',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Obtendo amostras
def extract_data(generator, num_samples):
    x_data = []
    y_data = []
    for _ in range(num_samples // generator.batch_size):
        x_batch, y_batch = next(generator)
        x_data.append(x_batch)
        y_data.append(y_batch)
    x_data = np.vstack(x_data)
    y_data = np.vstack(y_data)
    return x_data, y_data

# Ajustar o número total de amostras
num_train_samples = len(train_generator.filenames)
num_val_samples = len(val_generator.filenames)

x_train, y_train = extract_data(train_generator, num_train_samples)
x_val, y_val = extract_data(val_generator, num_val_samples)

kf = KFold(n_splits=5, shuffle=True, random_state=42)

best_val_accuracies = []
best_f1_scores = []
best_model_paths = []

for fold, (train_index, val_index) in enumerate(kf.split(x_train)):
    print(f"Treinando o fold {fold + 1}/{kf.n_splits}...")
    x_train_fold, x_val_fold = x_train[train_index], x_train[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    model = create_model(num_classes)
    
    # Treina o modelo
    model.fit(
        x_train_fold, 
        y_train_fold,
        validation_data=(x_val_fold, y_val_fold),
        epochs=num_epochs,
        callbacks=create_callbacks()
    )

    # Avalia o modelo
    val_loss, val_accuracy = model.evaluate(x_val_fold, y_val_fold, verbose=0)
    print(f'Fold {fold + 1} - Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}')

    # Plota matriz de confusão para o fold atual
    y_val_pred = model.predict(x_val_fold, batch_size=batch_size)
    y_val_pred_classes = np.argmax(y_val_pred, axis=1)
    y_val_true = np.argmax(y_val_fold, axis=1)
    confusion_mtx = confusion_matrix(y_val_true, y_val_pred_classes)
    plot_confusion_matrix(confusion_mtx, classes=['BCC', 'Melanoma', 'Desconhecido'], fold=f'Fold {fold + 1}')

    # Calcula o F1-score para o fold atual
    report = classification_report(
        y_val_true, 
        y_val_pred_classes, 
        labels=[0, 1, 2],  # Adiciona as classes esperadas
        target_names=['BCC', 'Melanoma', 'Desconhecido'],
        output_dict=True
    )
    f1_score = (report['BCC']['f1-score'] + report['Melanoma']['f1-score'] + report['Desconhecido']['f1-score']) / 3
    best_val_accuracies.append(val_accuracy)
    best_f1_scores.append(f1_score)
    
    # Salva o melhor modelo do fold
    best_model_path = f'/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/skin_model_keras/epoch_all/model_fold_{fold + 1}.keras'
    best_model_paths.append(best_model_path)
    model.save(best_model_path)
    
    import gc
    gc.collect()

# Exibe os resultados após todos os folds
print("Médias dos folds:")
print(f"Acurácia média: {np.mean(best_val_accuracies):.4f}")
print(f"F1-score médio: {np.mean(best_f1_scores):.4f}")

# Melhor modelo geral
best_fold_index = np.argmax(best_f1_scores)
print(f"O melhor modelo é o do fold {best_fold_index + 1} com F1-score: {best_f1_scores[best_fold_index]:.4f}")
```

```python
import tensorflow as tf

# Caminho do modelo salvo em formato Keras
model_path = '/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/01-mobileNetV2/fold_all/model_fold_3.keras'

# Carregar o modelo Keras
model = tf.keras.models.load_model(model_path)

# Criar um conversor TFLite
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# Converter o modelo para TFLite
tflite_model = converter.convert()

# Salvar o modelo convertido em um arquivo .tflite
tflite_model_path = '/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/01-mobileNetV2/fold_all/model_fold_1.tflite'
with open(tflite_model_path, 'wb') as f:
    f.write(tflite_model)

print(f"Modelo TFLite salvo em: {tflite_model_path}")
```

```python
# Criar o conversor TFLite com otimização
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]  # Quantização padrão

# Converter e salvar o modelo otimizado
tflite_quant_model = converter.convert()

# Salvar o modelo quantizado em um arquivo .tflite
tflite_quant_model_path = '/home/cristiano/www/skin_cancer_prediction/ISIC_2020/coding/01-mobileNetV2/fold_all/model_fold_1_quant.tflite'
with open(tflite_quant_model_path, 'wb') as f:
    f.write(tflite_quant_model)

print(f"Modelo TFLite quantizado salvo em: {tflite_quant_model_path}")
```
